```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
features = [1,2,3,4]
distances1 = [10,15,12,4]
distances2 = [5,10,11,5]
df = pd.DataFrame(list(zip(features, distances1, distances2)), columns =['Feature','Distance1','Distance2'])
print(df)

df2 = pd.melt(df, id_vars = "Feature", var_name = "Measure", value_name = "Distance")
print(df2)
#sns.histplot(x=x,y=y)
```

```python
fig, ax1 = plt.subplots(figsize=(10, 10))
sns.barplot(x="Feature", y="Distance", hue = 'Measure',data=df2,palette="pastel")
```
